```python
import pandas as pd
import numpy as np
```

```python
data_preprocessed = pd.read_csv('test1_preprocessed.csv')
```

```python
data_preprocessed.tail()
```

```python
#ターゲットの作成
data_preprocessed['Absenteeism Time in Hours'].median()
absent_median = data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > absent_median, 1, 0)
```

```python
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
data_preprocessed.tail()
```

```python
#データの調整
targets.sum() / targets.shape[0]
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours', 'Daily Work Load Average', 'Distance to Work'], axis=1)
```

```python
data_with_targets.head()
```

```python
#入力データの作成
data_with_targets.shape
```

```python
data_with_targets.iloc[:, 0:14]
```

```python
data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs.tail()
```

```python
#データの標準化
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
#データの分割
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size=0.8, random_state=20)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#モデルの作成
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

```python
#正確性の導出
model_outputs = reg.predict(x_train)
```

```python
np.sum((model_outputs == y_train) / model_outputs.shape[0])
```

```python
#切片の係数
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature name'], data = feature_name)
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
```

```python
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
#係数の解釈
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
summary_table
```

```python
#モデルのテスト
reg.score(x_test, y_test)
```

```python
predicted_proba = reg.predict_proba(x_test)
predicted_proba
```
